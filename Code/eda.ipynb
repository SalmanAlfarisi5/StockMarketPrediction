```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
merged_df = pd.read_csv('../Data/merged.csv')
merged_df.head()

# Change Date column to datetime format
merged_df['Date'] = pd.to_datetime(merged_df['Date'], format='mixed')
```

```python
# Let's pick one company AAPL and visualize its stock price over time
aapl = merged_df[merged_df['Symbol'] == 'AAPL'].sort_values('Date')

aapl_counts = aapl['Date'].value_counts()

# check whether there's no duplicates
print(aapl_counts[aapl_counts > 1])

print(aapl_counts.index.to_list()) 
```

```python

# Plot Close price over time
plt.figure(figsize=(12, 6))
plt.plot(aapl['Date'], aapl['Close'], linestyle='-')
plt.title('AAPL Stock Price (Close) Over Time')
plt.xlabel('Year')
plt.ylabel('Close Price')

# Format x-axis to show only one tick per year
years = pd.to_datetime(['2020-01-01', '2021-01-01', '2022-01-01'])
plt.xticks(ticks=years, labels=[str(year.year) for year in years])

plt.grid(True)
plt.tight_layout()
plt.show()
```
